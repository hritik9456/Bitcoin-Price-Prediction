# Data Preprocessing and loading

```python
from alpha_vantage.cryptocurrencies import CryptoCurrencies
import matplotlib.pyplot as plt
## getting data from alpha_vantage
cc = CryptoCurrencies(key='YI9Q3UBLBU4M6EW6', output_format='pandas')
data, meta_data = cc.get_digital_currency_daily(symbol='BTC', market='INR')
data['4a. close (INR)'].plot()
plt.tight_layout()
plt.title('Daily closing value for bitcoin (BTC)')
plt.grid()
plt.show()
```

```python
data.shape        
data.head(10)
#data.index
```

```python
useful_column = ['1a. open (INR)','2a. high (INR)','3a. low (INR)','4a. close (INR)']     # data used for predictions
my_data = data[useful_column]                      ## data to be converted into training and test
train_stat = my_data.describe()  
train_stat = train_stat.transpose()
train_stat.describe()# to print summary of the dataset
```

```python
def norm(x):                 
    """
    Normalization of data is done using this function
    """
    return (x-train_stat['mean'])/train_stat['std']
```

```python
train_X = my_data[120:]                     # training data
test_X = my_data[0:120]                     # test data
test_X.head()
```

```python
X_train = train_X[0:len(train_X)-1]                  # training data to be fed into model
Y_train = train_X[1:len(train_X)] # training label
X_test = test_X[0:len(test_X)-1]                     # test data to be fed into model
Y_test = test_X[1:len(test_X)]    # test label
### normalizing data
X_test = norm(X_test)
Y_test = norm(Y_test)
Y_test_further = Y_test.copy()
X_train = norm(X_train)
Y_train = norm(Y_train)
```

```python
## function to calculate R2 score
def R2(y_true, y_pred):
    s_res =  np.sum(np.square(y_true - y_pred)) 
    s_tot = np.sum(np.square(y_true - np.mean(y_true))) 
    return ( 1 - s_res/(s_tot + bck.epsilon()) )
```

```python
## function to retrieve data back(normalized data) 
def back_to_original(x):
    return (x*train_stat['std']['4a. close (INR)']+train_stat['mean']['4a. close (INR)'])
```
